```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./synthetic_financial_data.csv')
```

```python
data
```

```python
data.info()
```

```python
data['transaction_time'] = pd.to_datetime(data['transaction_time'])
```

```python
data[['amount', 'customer_age']].describe()
```

```python
data_fraudulenta = data['is_fraudulent'].value_counts().reset_index()
data_fraudulenta
```

```python
plt.title("Cantidad de data fraudulenta vs No Fraudulenta")
plt.bar(["Fraudulento", "No fraudulento"], data_fraudulenta['count'], color = 'skyblue')
plt.xlabel("Tipo")
plt.ylabel("Cantidad")
```

```python
data
```

```python
df_data_fraudulenta = data[(data['is_fraudulent'] == 1)]
df_data_fraudulenta
```

```python
tipo_comercio_fraude = df_data_fraudulenta.groupby('is_fraudulent')['purchase_category'].value_counts().reset_index()
tipo_comercio_fraude
```

```python
tipo_comercio_fraude = df_data_fraudulenta['purchase_category'].value_counts().reset_index()
tipo_comercio_fraude
```

```python
plt.pie(tipo_comercio_fraude['count'], labels=tipo_comercio_fraude['purchase_category'], autopct='%1d%%', startangle= 140)
plt.legend(tipo_comercio_fraude['purchase_category'], title="Categories", loc="center", bbox_to_anchor=(1.1, 0, 0.5, 1))
```

```python
tipo_tarjeta_fraude = df_data_fraudulenta['card_type'].value_counts()
tipo_tarjeta_fraude
```

```python
plt.title("Tipo de tarjeta usada en el fraude")
plt.bar(tipo_tarjeta_fraude.index, tipo_tarjeta_fraude.values, color = 'skyblue')
plt.xlabel("Tipo de tarjeta")
plt.ylabel("Cantidad")
plt.xticks(rotation = 45, ha = 'right')
```
