```python
merged_df = pd.merge(data_final, data_enero[['AG DIG 6', 'VALOR']], left_on='cuenta', right_on='AG DIG 6', how='left')
```

```python
merged_df
```

```python
merged_df['enero'] = merged_df['VALOR'] / 1000000
```

```python
merged_df = merged_df.drop(columns=['AG DIG 6', 'VALOR'])
```

```python
merged_df
```

```python
columnas_sumar = ["agosto", "septiembre", "octubre", "noviembre", "diciembre", "enero"]
```

```python
def calcular_acumulado(row):
    if row['Categoria'] in ['Pagos', 'Servicios']:
        return row[columnas_sumar].mean(skipna=True)
    else:
        return row[columnas_sumar].sum(skipna=True)
```

```python
merged_df['Acumulado 2024'] = merged_df.apply(calcular_acumulado, axis=1)
```
